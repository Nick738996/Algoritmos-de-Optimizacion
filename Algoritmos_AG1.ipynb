##Torres de Hanoi con Divide y vencerás

```
# TÉCNICA: DIVIDE Y VENCERÁS
# Problema: Torres de Hanói

# OBJETIVO: Mover las fichas de la torre 1 a la 3 usando la 2 como auxiliar.
# COMPLEJIDAD: O(2^n). Es exponencial. Con 3 fichas es rápido, pero con 64 sería eterno.

def Torres_Hanoi(N, desde, hasta):
  # Caso base: Si queda una sola ficha, la movemos directo
  if N == 1:
    print("Lleva la ficha ", desde, " hasta ", hasta)

  else:
    # 1. DIVIDE: Movemos N-1 fichas al poste auxiliar (Pivote).
    # Truco matemático: 1+2+3=6. Por tanto, (6 - origen - destino) nos da el poste libre.
    Torres_Hanoi(N-1, desde, 6-desde-hasta)

    # 2. VENCE: Movemos la ficha grande al destino
    print("Lleva la ficha ", desde, " hasta ", hasta)

    # 3. COMBINA: Movemos las N-1 fichas del auxiliar al destino
    Torres_Hanoi(N-1, 6-desde-hasta, hasta)

# PRUEBA
# Con 3 fichas deberían salir 7 movimientos (2^3 - 1)
print("--- HANOI CON 3 FICHAS ---")
Torres_Hanoi(3, 1, 3)
```

```
# Sucesión_de_Fibonacci
# https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci

# Cálculo del término n-simo de la sucesión de Fibonacci
# Nota: Esta implementación es recursiva pura.
# Complejidad: O(2^n) = Exponencial.
# Problema: Es muy ineficiente para N grandes porque recalcula los mismos subproblemas
# una y otra vez. (Ideal para explicar por qué necesitamos Programación Dinámica).

def Fibonacci(N:int):
  if N < 2:
    return 1
  else:
    return Fibonacci(N-1) + Fibonacci(N-2)

# Prueba
print("Fibonacci(5):", Fibonacci(5))

# Si intentaramos Fibonacci(40) o más, veriamos que el ordenador se queda pensando por un largo tiempo
```

##Devolución de cambio por técnica voraz

```
# TÉCNICA: ALGORITMOS VORACES (GREEDY)
# Problema: Devolución de cambio

# Lógica: En cada iteración elegimos la moneda más grande posible.
# Complejidad: O(N) donde N es el número de tipos de monedas.
# Requisito: El sistema de monedas (SM) debe estar ordenado de mayor a menor.

def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   # Inicializamos el array solución con ceros
  ValorAcumulado = 0

  for i, valor in enumerate(SM):
    # Calculamos cuántas monedas de este valor caben
    monedas = (N - ValorAcumulado) // valor

    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas * valor

    # Si ya hemos devuelto todo el cambio, terminamos
    if ValorAcumulado == N:
      return SOLUCION

  return SOLUCION

# PRUEBA
print("Solución Voraz para 15:", cambio_monedas(15, [25, 10, 5, 1]))
```

```
# TÉCNICA: VUELTA ATRÁS (BACKTRACKING)
# Problema: Las N-Reinas

# Con N=8, el algoritmo encontrará 92 soluciones distintas.
# Imprimimos solo la lista numérica para no saturar la pantalla con 92 tableros.

def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")
  print("\n")

def es_prometedora(SOLUCION, etapa):
  # Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    # Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]): return False
  return True

def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        # Imprimimos la solución numérica (Ej: [1, 5, 8...])
        print(solucion)
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

# PRUEBA: 8 Reinas (Generará 92 líneas de soluciones tambien podriamos reducir a 4 para no generar tantas soluciones)
reinas(8)
```

## Viaje por el rio. Programación dinámica

```
# TÉCNICA: PROGRAMACIÓN DINÁMICA
# Problema: Viaje por el río (Cálculo de rutas óptimas en grafos)

# Estrategia: Construimos una tabla (PRECIOS) donde almacenamos el coste mínimo
#             para ir de un nodo 'i' a un nodo 'j'.
#             Usamos los resultados de los caminos cortos para calcular los largos.
# Complejidad: O(N^3) -> Tres bucles anidados.

TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]

def Precios(TARIFAS):
  # Total de Nodos (7 en este caso)
  N = len(TARIFAS[0])

  # Inicialización de la tabla de precios con valor infinito (9999)
  # y tabla de rutas vacía
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  # Bucle principal: llenamos la tabla
  for i in range(0,N-1):
    RUTA[i][i] = i             # Ir de i a i es quedarse en i
    PRECIOS[i][i] = 0          # Coste 0

    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      # Bucle intermedio (k): Buscamos si pasar por un nodo 'k' es más barato
      # que ir directo o por la ruta que conocíamos antes.
      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          # Guardamos el nodo intermedio clave
        PRECIOS[i][j] = MIN

  return PRECIOS, RUTA

# EJECUCIÓN
PRECIOS, RUTA = Precios(TARIFAS)

print("--- MATRIZ DE PRECIOS ÓPTIMOS ---")
# Imprimimos fila a fila para que se vea como una matriz
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\n--- MATRIZ DE RUTAS ---")
for i in range(len(TARIFAS)):
  print(RUTA[i])

# Función recursiva para reconstruir el camino leyendo la matriz RUTA
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    return desde
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\n--- RUTA MÁS BARATA DE 0 A 6 ---")
# Nota: La función devuelve los nodos intermedios.
# El coste total lo puedes ver en PRECIOS[0][6]
ruta_final = calcular_ruta(RUTA, 0, 6) + ",6"
print(f"Ruta: {ruta_final}")
print(f"Coste total: {PRECIOS[0][6]}")
```
